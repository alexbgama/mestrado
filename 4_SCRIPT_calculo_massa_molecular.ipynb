### Este script calcula a massa molecular das moléculas usando SMILES como entrada e exporta uma tabela excel SMILES x MM

1. Ter bibliotecas instaladas: pandas e rdkit 

```python
# !pip install rdkit pandas
```

2. Importar bibliotecas

```python
from rdkit import Chem
from rdkit.Chem.Descriptors import ExactMolWt
import pandas as pd
```

3. Definição da lista de SMILES

```python
# Lista de SMILES das moléculas a serem analisadas

smiles_list = ['CC(C)(C)C1=CC(=C(C=C1O)C(C)(C)C)O', 'CC(C)C1=CC=C(CC2CCCC(CC3=CC=C(C=C3)C(C)C)C2=O)C=C1', 'ClC1=CC(Cl)=C(CC2CCCC(CC3=C(Cl)C=C(Cl)C=C3)C2=O)C=C1',
               'ClC1=CC=C(CC2CCCC(CC3=CC=C(Cl)C=C3)C2=O)C=C1', 'CCCCO', 'COc1ccccc1C=CC(=O)c1ccccc1NS(=O)(=O)c1ccccc1',
               'Brc1ccccc1C=CC(=O)c1ccccc1NS(=O)(=O)c1ccccc1', 'Clc1ccccc1C=CC(=O)c1ccccc1NS(=O)(=O)c1ccccc1', 'Clc1ccc(cc1)C=CC(=O)c1ccccc1NS(=O)(=O)c1ccccc1',
               'O=C(C=Cc1cccc(c1)N(=O)=O)c1ccccc1NS(=O)(=O)c1ccccc1', 'COC1=C(C=CC(=C1)/C=C/C(=O)CC(=O)/C=C/C2=CC(=C(C=C2)O)OC)O',
               'CCO', 'BrC1=CC=C(CCC(=O)C2=CC=CC(OCC(=O)C3=CC=CC=C3)=C2)C=C1', 'O=C(COC1=CC=CC(=C1)C(=O)\C=C\C1=CC=CC=C1)C1=CC=CC=C1',
               '[H]\C(=C(\[H])C1=C([H])C([H])=C(Cl)C([H])=C1[H])C(=O)C1=C([H])C(OC([H])([H])C(=O)C2=C([H])C([H])=C([H])C([H])=C2[H])=C([H])C([H])=C1[H]',
               'BrC1=CC=C(CCC(=O)C2=CC=C(OCC(=O)C3=CC=CC=C3)C=C2)C=C1', '[H]N(C1=C([H])C([H])=C([H])C([H])=C1C(=O)C(\[H])=C(/[H])C1=C([H])C([H])=C(OC([H])([H])[H])C([H])=C1[H])S(=O)(=O)C1=C([H])C([H])=C([H])C([H])=C1[H]',
               '[H]N(C1=C([H])C([H])=C([H])C([H])=C1C(=O)C(\[H])=C(/[H])C1=C([H])C([H])=C(Br)C([H])=C1[H])S(=O)(=O)C1=C([H])C([H])=C([H])C([H])=C1[H]',
               'Br[Br]C1=CC=C(CC2C(N(C3=C(C=CC=C3)C2=O)S(=O)(=O)C2=CC=CC=C2)C2=CC=C(Br)C=C2)C=C1', 'CC(C)(C)c1cc(O)ccc1O'
               ]
```

4. Cálculo da massa molecular exata

```python
# Lista para armazenar os dados resultantes
dados = []

# Loop para converter os SMILES e calcular massa exata
for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    
    if mol:
        massa_exata = ExactMolWt(mol)
        dados.append({
            'SMILES': smiles,
            'Massa Molecular Exata (g/mol)': round(massa_exata, 4)
        })
    else:
        dados.append({
            'SMILES': smiles,
            'Massa Molecular Exata (g/mol)': 'Inválido'
        })
```

5.  Montar a tabela final

```python
# Cria DataFrame com os dados coletados
df = pd.DataFrame(dados)
df
```

6. Exportar para excel

```python
# Exporta o DataFrame como "massa_molecular.xlsx"

df.to_excel("massa_molecular.xlsx", index=False)

# Mensagem de confirmação

print("Arquivo 'massa_molecular.xlsx' exportado com sucesso!")
```
